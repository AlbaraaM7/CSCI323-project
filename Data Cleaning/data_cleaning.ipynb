```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from category_encoders import BinaryEncoder
```

```python
raw_data_path= '../Data/Raw Data/'
clean_data_path= '../Data/Clean Data/'
```

```python
raw_data_file=raw_data_path+'raw_data.csv'
df=pd.read_csv(raw_data_file)
df.head()
```

```python
df.info()
```

```python
#Rename columns for easier navigation
column_names={
    'TRANSACTION_NUMBER':'transaction_num',
    'INSTANCE_DATE':'instance_date',
    'GROUP_EN':'group',
    'PROCEDURE_EN':'procedure',
    'IS_OFFPLAN_EN':'is_offplan',
    'IS_FREE_HOLD_EN':'is_free_hold',
    'USAGE_EN':'usage',
    'AREA_EN':'area',
    'PROP_TYPE_EN':'prop_type',
    'PROP_SB_TYPE_EN':'prop_sb_type',
    'TRANS_VALUE':'trans_value',
    'PROCEDURE_AREA':'procedure_area',
    'ACTUAL_AREA':'actual_area',
    'ROOMS_EN':'rooms',
    'PARKING':'parking',
    'NEAREST_METRO_EN':'nearest_metro',
    'NEAREST_MALL_EN':'nearest_mall',
    'NEAREST_LANDMARK_EN':'nearest_landmark',
    'TOTAL_BUYER':'total_buyer',
    'TOTAL_SELLER':'total_seller',
    'MASTER_PROJECT_EN':'master_project',
    'PROJECT_EN':'project'
}
df.rename(columns=column_names,inplace=True)
df
```

```python
#Check number of unique values in each column
print("Number of unique values in each column:")
for x in df.columns:
    print(f"{x}: {len(df[x].unique())}")
print("\n\n\nNumber of null values in each column:")
for x in df.columns:
    print(f"{x}: {df[x].isnull().sum()}")
```

```python
#Drop useless columns
df_original=df.copy()
df=df.drop(columns=['transaction_num','total_buyer','total_seller','master_project'])
df
```

```python
# Fix and separate instance date
df['instance_date']=pd.to_datetime(df['instance_date'])

df['date_year']        = df['instance_date'].dt.year
df['date_month']       = df['instance_date'].dt.month
df['date_day']         = df['instance_date'].dt.day
df['date_hour']        = df['instance_date'].dt.hourg
df['date_day_of_week'] = df['instance_date'].dt.dayofweek
#Get rid of original column since its useless
df=df.drop(columns=['instance_date'])
```
